```python
import numpy as np
import matplotlib.pyplot as plt
import os

# --- Ορισμός παραμέτρων ---
#beta = 0.4974 #delta
beta=0.8589 #omikron
epsilon_mask = 0.79
#k = 0.1726 #delta
k=0.6968 #omikron
n_a = 0.76
p = 0.25
l = 0.93

sigma_u = 1/5.8
sigma_alpha = 1
gamma_a = 1/6
gamma_i = 1/6
sigma_h = 1/4
mu = 1/(82*365)
kl = k * l

#f1 = 0.3647 #delta
f1=0.1921 #omikron
#f2 = 0.038 #delta
f2=0.0218 #omikron
f2_prime = f2 / 10

# --- Συναρτήσεις για υπολογισμό στοιχείων πίνακα ---
def V_inv_11():
    return 1 / (sigma_u + mu)

def V_inv_22():
    return 1 / (sigma_u + mu)

def V_inv_33():
    return 1 / (gamma_a + kl * sigma_alpha + mu)

def V_inv_44():
    return 1 / (gamma_a + kl * sigma_alpha + mu)

def V_inv_55():
    return 1 / (f2 * sigma_h + (1 - f2) * gamma_i + mu)

def V_inv_66():
    return 1 / (f2_prime * sigma_h + (1 - f2_prime) * gamma_i + mu)

# --- Συνάρτηση υπολογισμού του R0 ---
def compute_R0(p_mask, p_dist, e_vac):
    c = beta * (1 - p_dist) * (1 - epsilon_mask * p_mask)

    K11 = c * V_inv_11()
    K12 = c * (1 - k) * n_a * V_inv_33()
    K13 = c * p * V_inv_55()

    K21 = c * (1 - e_vac) * V_inv_22()
    K22 = c * (1 - e_vac) * (1 - k) * n_a * V_inv_44()
    K23 = c * (1 - e_vac) * p * V_inv_66()

    a = K11**2 + K12**2 + K13**2
    d = K21**2 + K22**2 + K23**2
    b = K11 * K21 + K12 * K22 + K13 * K23

    R0 = np.sqrt((a + d)/2 + 0.5 * np.sqrt((a - d)**2 + 4 * b**2))

    return R0


# --- Τιμές για περίοδο Δέλτα (εκτίμηση, μπορείς να τις τροποποιήσεις) ---
p_mask_delta = 0.6344 # ποσοστό πληθυσμού που φορά μάσκα 0.4515
p_dist_delta = 0.6457        # ποσοστό κοινωνικής αποστασιοποίησης 0.622
e_vac_delta = 0.75        # ποσοστό εμβολιαστικής κάλυψης

# --- Υπολογισμός Rt για Δέλτα ---
Rt_delta = compute_R0(p_mask_delta, p_dist_delta, e_vac_delta)
print(f"Rt για Δέλτα (με μέτρα): {Rt_delta:.3f}")

# --- Υπολογισμός R0 (χωρίς κανένα μέτρο) ---
R0 = compute_R0(0, 0, 0)
print(f"R0 (χωρίς κανένα μέτρο): {R0:.3f}")
```
