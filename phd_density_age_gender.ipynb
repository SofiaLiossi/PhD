```python
import scipy.stats as stats

# Mann-Whitney U test for difference between men and women
ct_men = df[df['gender'] == 'man']['ct']
ct_women = df[df['gender'] == 'woman']['ct']
u_statistic, p_value_mannwhitney = stats.mannwhitneyu(ct_men, ct_women, alternative='two-sided')

print(f"Mann-Whitney U test results for gender difference:")
print(f"U statistic = {u_statistic:.2f}, p-value = {p_value_mannwhitney:.4f}")

# Kruskal-Wallis H test for difference across age categories
age_groups = [df[df['age group'] == age]['ct'] for age in df['age group'].unique()]
h_statistic, p_value_kruskal = stats.kruskal(*age_groups)

print(f"\nKruskal-Wallis H test results for age group difference:")
print(f"H statistic = {h_statistic:.2f}, p-value = {p_value_kruskal:.4f}")
```

```python
import scipy.stats as stats

# One-way ANOVA for gender difference (equivalent to t-test for two groups)
f_statistic_gender, p_value_gender = stats.f_oneway(ct_men, ct_women)

print(f"One-way ANOVA results for gender difference:")
print(f"F statistic = {f_statistic_gender:.2f}, p-value = {p_value_gender:.4f}")

# One-way ANOVA for age group differences
age_groups_anova = [df[df['age group'] == age]['ct'] for age in df['age group'].unique()]
f_statistic_age, p_value_age = stats.f_oneway(*age_groups_anova)

print(f"\nOne-way ANOVA results for age group differences:")
print(f"F statistic = {f_statistic_age:.2f}, p-value = {p_value_age:.4f}")
```
